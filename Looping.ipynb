#looping--------->repeat a block of code multiple times

finite looping------------>we know the number of iteration #for loop----sequence of items.
infinite looping------------>we dont know number of iteration # while loop..



#for loop syntax

for variable in iteration:
    block of code

```python
list1=["apple","banana","cherry","veggies","corn"]

for i in list1:
    print(i)
```

```python
dict1={"name":"sri","age":24,"course":"ai"}

for i,j in dict1.items():
    print(i,j)
```

```python
#range()

for i in range(6):
    print(i,end=" ")
```

```python
for i in range(1,100,2):
    print(i)
```

```python
for i in range(1,100):
    if i%2==0:
        print(i)
```

```python
row=3

for i in range(1,row+1):
    print("*"*i)
```

```python
table_no=int(input())

for i in range(1,21):
    print(f"{table_no} x {i} ={table_no*i}")
```

```python
n=int(input())
sum=0
for i in range(1,n+1):
    sum=sum+i
    print(sum)
```

```python
#while loop
# syntax
# while condition:
#     block of code

i=0

while i<=5:
    print(i)
    i+=1   #i=i+i
```

```python
correct_password="admin@123"
password=input("Enter the password")

while password!=correct_password:
    print("Wrong password try again")
    password=input("Enter the password")

print("access granted")
```

```python
#looping control statments.

break----->exit the loop immediatly when encountered
continue----->skips the current the iteration
pass--------->it acts as placeholder for future code reference
```

```python
for i in range(1,10):
    if i==3:
        break
    print(i)
```

```python
for i in range(1,10):
    if i==3:
        continue
    print(i)
```

```python
for i in range(1,10):
    if i==3:
       pass
    print(i)
```
