# inheritance---------->one class inherits properties,methods from another class

# why inheritance?
code reusability

types of inheritance:

```python
#single inheritance:

class animal:
    def show(self):
        print("Animals...........")

class dog(animal):
    def sound(self):
        print("Dogs makes sound")
    def show(self):
        super().show()
        print("Dogs shows")
        animal.show(self)


d1=dog()
d1.sound()
d1.show()
```

```python
#multilevel inheritance

class grandparent:
    def work(self):
        print(f"Grandpa reading news paper")
class parent(grandparent):
    def display(self,name):
        print(f"parent name {name}")
    def work(self):
        print(f"parents working in a company")
        super().work()

class child(parent):
    def show(self):
        print(f"child.......")
    def work(self):
        print(f"child studying")
        super().work()

c1=child()
c1.display("priya")
c1.work()
```

```python
#multiple

class father:
    def display(self,name,job):
        print(f"Name:{name}")
        print(f"Job:{job}")

class mother:
    def show(self,name,job):
        print(f"Name:{name}")
        print(f"job:{job}")

class son(father,mother):
    def display(self,name,job):
        print(f"name:{name}")
        print(f"job:{job}")

s1=son()
s1.display("sri","ai engineer")

s1.show("priya","housewi)
```

```python
#single

class animal:
    def show(self):
        print(f"Animals sleeping")

class dog(animal):
    def sound(self):
        print(f"dog makes sound")


d1=dog()
d1.show()
d1.sound()
```

```python
#multilevel
class animal:
    def show(self):
        print(f"Animals sleeping")

class dog(animal):
    def sound(self):
        print(f"dogs makes sound")

class pug(dog):
    def sound(self):
        print(f"sounds like haha")
        super().sound()
        dog.sound(self)

        
c1=pug()

c1.show()
c1.sound()
```

```python
#multiple

class father:
    def work(self):
        print(f"father working in a company")
        mother.work(self)

class mother:
    def work(self):
        print(f"mother cooking")
        

class son(father,mother):
    def work(self):
        print(f"son studying")
        super().work()

s1=son()
s1.work()
```

```python
#hierchical inheritance

class animal:
    def sound(self):
        print("animals.........")

class dog(animal):
    pass
class elephant(animal):
    pass
class cat(animal):
    pass
class tiger(animal):
    pass

t1=tiger()
t1.sound()

d1=dog()
d1.sound()

c1=cat()
c1.sound()

e1=elephant()
e1.sound()
```

```python
import math

# from abc import Abc,abstractmethod
class shape:
    # @abstractmethod
    def area(self):
        pass

class rectangle(shape):
    def __init__(self,length,width):
        self.length=length
        self.width=width
    def area(self):
        return self.length*self.width


class circle(shape):
    def __init__(self,radius):
        self.radius=radius

    def area(self):
        return math.pi*self.radius*self.radius

class triangle(shape):
    def __init__(self,b,h):
        self.b=b
        self.h=h
    def area(self):
        return 0.5*self.b*self.h


shape=[circle(4)]

for shapes in shape:
    print(shapes.area())

# r=rectangle(10,2)
# print(r.area())

# c=circle(4)
# print(c.area())

# t=triangle(7,6)
# t.area()
```
