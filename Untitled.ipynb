```python
def student(name,age,address):    #keyword arguments
    return f"name: {name}, age: {age}, address:{address}"
student(age=24,name="sri",address="cbe")
```

```python
def numbers(*num):   #variable length positional arguments
    print(sum(num))
numbers(10,90,89,78,67,89,90,78,67,56,45,13,12,12)
```

```python
def kwargs(**kwargs):   #varaible length keyword arguments
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
#variable types
global variable----->outside of any function
local variable------->within the function
nonlocal variable-------->nested functions
```

```python
y=10
def access():
    global 
    
    
    
    print(y)

print(y)
access()
```

```python
a=10

def value():
    global a
    a=15
    print(a)

value()
print(a)
```

```python
x=10
def outer():
    x="outer value"
    def inner():
        nonlocal x
        x="inner value"

    inner()
    print(x)

outer()
```

```python
x = 5  # global variable

def change_global():
    global x
    x = 10  # modifies the global variable

change_global()
print(x)  # Output: 10
```
