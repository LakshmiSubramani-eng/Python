```python
number=int(input("Enter the number"))
duplicate=number
reverse=0

while number>0:
    digit=number%10
    reverse=reverse*10+digit
    number=number//10

print(reverse)
if reverse==duplicate:
    print("its palidrome")
else:
    print("its not a palidrome")
```

```python
#looping control statments--------->break,continue, pass

break--->break the code
continue--->skip the current iterations
pass----->its does nothing, acts as a place holder
```

```python
for i in range(1,10):
    if i==5:
        break
    print(i)
```

```python
for fruits in (["apple","banana","cherry"]):
    if fruits=="banana":
        break
    print(fruits)
```

```python
for i in range(1,10):
    if i==5:
        continue
    print(i)
```

```python
for i in range(1,10):
    if i==5:
        pass
    print(i)
```
