```python
#operators----------------->its like symbols

1.arithemetic operators
2.assigment operators
3.comparison or relational operators
4.logical operators
5.bitwise operators
6.identity operators
7.membership operators
```

```python
#arithmetic operators-------> +,-,*,/,//,**,%

a=int(input("Enter the number1"))
b=int(input("Enter the number2"))

print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a//b)
print(a**b)
print(a%b)
                  
```

```python
#comparison operator------> >,<,<=,>=,!=,==

a=int(input("Enter the number1"))
b=int(input("Enter the number2"))

print(a<b)
print(a>b)
print(a>=b)
print(a<=b)
print(a==b)
print(a!=b)

           
```

```python
#assigment operators-->+=,-=,*=,

x=10
x+=20 #x=x+20
x*=10  #x=x*10
print(x)


y=10
y**=x   #y=y**x------>10**300
print(y)
```

```python
#logical operators------->and,or,not

a=10
b=12
print(a>b and a!=b)
print(a>b or a!=b)
print(not(a>b))
```

```python
#membership operators--------->in, not in

a=["Apple","banana","cherry","guva"]

print("cherry" in a)
print("pomeogranate" not in a)
```

```python
#identity operators------->is ,isnot

x=["apple","banana"]
y=["apple","banana"]

z=x

print(x is y)
print(z is x)
print(x==y)
```

```python
#bitwise operator

a=10
b=12

print(a&b)
print(a|b)
print(a^b)
print(~a)  #-(a+1)
```
